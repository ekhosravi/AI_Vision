```python
def show_multi_img (rows,columns,img_names,titles = None,vmin = 0,vmax = 255):
    fig = plt.figure(figsize=(15,17),dpi=100)
    for i in range(len(img_names)):
        fig.add_subplot(rows,columns,i+1)
        plt.imshow(img_names[i],cmap='gray',vmin=vmin,vmax=vmax)
        plt.axis('off')
        if titles is not None:
            plt.title(str(i+1) + '. ' + titles[i])
    plt.show()
```

```python
def show_img_plt(img, c_map ='gray', fig_hight=8, fig_width=6, vmin=0, vmax=255):
    plt.figure(figsize=(fig_hight, fig_width), dpi=100)
    plt.imshow(img, cmap=c_map, vmin=vmin, vmax=vmax)
```

```python
#read Image
img = cv2.imread(r'E:\TA Filoger\Exercise\DIP\DIP19\3_ex_19\img\input.jpg')
```

```python
# Color Conversion
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# Denoising
img_blur = cv2.GaussianBlur(img_gray,(3,3),0)
```

```python
show_img_plt(img_blur)
```

```python
#Gradient
soblex = cv2.Sobel(src = img_blur,ddepth = -1,dx = 1,dy = 0,ksize = 5)
sobley = cv2.Sobel(src = img_blur,ddepth = -1,dx = 0,dy = 1,ksize = 5)
```

```python
show_multi_img(1,2,(soblex,sobley),titles=['Soble X','SobelY'])
```

```python
gradient = cv2.addWeighted(np.abs(soblex),0.5,np.abs(sobley),0.5,0)
```

```python
show_img_plt(gradient)
```

```python
laplacian = cv2.Laplacian(img_blur,-1)
```

```python
img_out = cv2.imread(r'E:\TA Filoger\Exercise\DIP\DIP19\3_ex_19\img\output.jpg')
```

```python
show_multi_img(1,2,(img_out,laplacian),titles=["Out put","Laplacian"])
```

```python
canny = cv2.Canny(image = img_blur,threshold1 = 245,threshold2 = 250)
```

```python
show_multi_img(1,2,(img_out,canny),titles=['Out put','Canny'])
```
